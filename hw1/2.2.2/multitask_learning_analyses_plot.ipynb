```python
## https://matplotlib.org/stable/api/markers_api.html
from matplotlib import pyplot as plt
import numpy as np
from numpy.polynomial.polynomial import polyfit

x = [2.61,2.82,2.77,3.98,4.08,3.89, 
     1.48, 1.99, 1.78,  2.15, 1.69, 2.65,  2.00, 1.25, 1.93,
     2.45,2.70, 
     2.41,2.00, 2.79, 2.25]
y = [-4.918,-4.898,-4.903,-4.905,-4.857,-4.825, 
     -5.164, -5.140, -5.141,  -5.118, -5.092, -5.088,  -5.158, -5.158, -5.143,
     -5.065,-5.036, 
     -5.055,-5.035, 
     -4.934, -4.960]

x = np.array(x)
y = np.array(y)

# Fit with polyfit
b, m = polyfit(x, y, 1)

x_ = np.arange(x.min(), x.max()+.1, .1)

plt.plot(x_, b + m * x_, linestyle='--', dashes=(5, 5), color='gray')

# scatter plot
plt.scatter([2.81], [-4.878], marker='o', label='bert-base-cased') ## 1 point
plt.scatter([2.61,2.82,2.77,3.98,4.08,3.89], 
            [-4.918,-4.898,-4.903,-4.905,-4.857,-4.825], marker='^', label='T1') ## 3 points
plt.scatter([1.48, 1.99, 1.78,  2.15, 1.69, 2.65,  2.00, 1.25, 1.93], 
            [-5.164, -5.140, -5.141,  -5.118, -5.092, -5.088,  -5.158, -5.158, -5.143], marker='v', label='T2') ## 3 points
plt.scatter([2.45,2.70],
            [-5.065,-5.036], marker='*', label='T1 -> T2') ## 1 point
plt.scatter([2.41,2.00], 
            [-5.055,-5.035], marker='d', label='T2 -> T1') ## 1 point
plt.scatter([2.79], 
            [-4.934], marker='p', label='T1 & T2 (FW)') ## 1 point
plt.scatter([2.25], 
            [-4.960], marker='h', label='T1 & T2 (DWA)') ## 1 point



plt.legend()
plt.xlabel('EM score (%)')
plt.ylabel('LEEP score')
```
